Plot temperature data from NCEP reanalysis data. xarray.open_dataset can open and decode a dataset from a file path or an OpenDAP URL.

The dataset that we will be using is here:
https://www.esrl.noaa.gov/psd/thredds/catalog/Datasets/ncep.reanalysis/surface/catalog.html

```python
import xarray as xr
file = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis/surface/air.sig995.2024.nc'

with xr.open_dataset(file) as ds:
    print(ds)

# Define surface air temperature ts
ts=ds.air-273.15
```

```python
# Plot the last time step
ts.isel(time=-1).plot()
```

Plot the second last time step

```python
#Plot temperature of a point 
lat_pvd=41.8
lon_pvd=-71.4+360
ts_prov=ts.sel(lat=lat_pvd,lon=lon_pvd , method='nearest')
ts_prov.plot()
```
